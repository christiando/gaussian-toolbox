```python
import pandas
import numpy
from matplotlib import pyplot
import sys
sys.path.append('../../timeseries/')
sys.path.append('../../src/')
import factors
import observation_models
import state_models
from ssm_em import StateSpaceEM
```

```python
df = pandas.read_csv('../../data/sunspots/monthly-sunspots.csv', parse_dates=True)
```

```python
df.plot(lw=0, marker='.')
```

```python
X = numpy.array([df['Sunspots'].to_numpy()]).T
X = numpy.hstack([X[1:],X[:-1]])
X_train = X[:1500]
X_test = X[1500:]
```

```python
Dz = 2
Dx = X.shape[1]
sm = state_models.LinearStateModel(Dz)
om = observation_models.LinearObservationModel(Dx, Dz)
#om.pca_init(X)
```

```python
ssm_em_lin = StateSpaceEM(X_train, observation_model=om, state_model=sm)
ssm_em_lin.run()
pyplot.plot(ssm_em_lin.llk_list[1:])
pyplot.xlabel('EM iterations')
pyplot.ylabel('Likelihood')
pyplot.show()
```

```python
Dz = 2
Dx = X.shape[1]
Du = 1
sm = state_models.LinearStateModel(Dz)
# Here we change the observation model.
om = observation_models.HCCovObservationModel(Dx, Dz, Du)
#om.pca_init(X_train)
ssm_em = StateSpaceEM(X_train, observation_model=om, state_model=sm)
ssm_em.run()
```

```python
p_x = ssm_em.compute_data_density()
p_x_lin = ssm_em_lin.compute_data_density()
pyplot.figure(figsize=(10,5))
for ix in range(1):
    pyplot.subplot(1,1,ix+1)
    pyplot.plot(X[:4000,ix],'k')
    x_min = p_x_lin.mu[:4000,ix] - 1.68 * numpy.sqrt(p_x_lin.Sigma[:4000,ix,ix])
    x_max = p_x_lin.mu[:4000,ix] + 1.68 * numpy.sqrt(p_x_lin.Sigma[:4000,ix,ix])
    pyplot.fill_between(range(x_min.shape[0]), x_min, x_max, color='C0', alpha=.5, zorder=8)
    x_min = p_x.mu[:4000,ix] - 1.68 * numpy.sqrt(p_x.Sigma[:4000,ix,ix])
    x_max = p_x.mu[:4000,ix] + 1.68 * numpy.sqrt(p_x.Sigma[:4000,ix,ix])
    pyplot.fill_between(range(x_min.shape[0]), x_min, x_max, color='C1', alpha=.5, zorder=9)
    pyplot.ylabel('x_%d' %(ix+1))
    pyplot.xlim([0,1500])
pyplot.xlabel('Time steps')
pyplot.show()
```
